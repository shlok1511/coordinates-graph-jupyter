```python
import matplotlib.pyplot as plt

def get_coordinates():
    print("Enter coordinates one by one in the format x,y (type 'done' to finish):")
    coords = []
    while True:
        user_input = input("Enter coordinate: ")
        if user_input.lower() == 'done':
            break
        try:
            x_str, y_str = user_input.split(',')
            x, y = float(x_str.strip()), float(y_str.strip())
            coords.append((x, y))
        except ValueError:
            print("Invalid format. Please enter coordinates as x,y")
    return coords

def plot_coordinates(coords):
    if not coords:
        print("No coordinates to plot.")
        return
    x_vals, y_vals = zip(*coords)
    plt.figure(figsize=(8, 6))
    plt.plot(x_vals, y_vals, 'bo-', label='Path')
    plt.title("User Input Coordinates")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.legend()
    plt.show()

if __name__ == "__main__":
    coordinates = get_coordinates()
    plot_coordinates(coordinates)
```
